Due to the existence of zeroes, we will impute very small rational numbers to these zero-values. On the other hand, we will encode 'B' with 0 and 'M' with 1 on the 'diagnosis' column to ensure compatibility with sklearn pipelines.

```python
# The following transformer will encode the 'diagnosis' columns with 0's and 1's.

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils.validation import check_is_fitted

class BinaryEncoder(BaseEstimator, TransformerMixin):
    
    def __init__(self):
        pass
    
    def fit(self,X,y=None):
        
        self.benign_index = X.index[X['diagnosis']=='B']
        self.malignant_index = X.index[X['diagnosis']=='M']
        
        return self
    
    def transform(self, X, y=None):
        
        X.loc[self.benign_index, 'diagnosis']=0
        X.loc[self.malignant_index, 'diagnosis']=1
        
        return X.values
```

```python
# The following transformer will replace zeroes with small rational numbers.

class noZeroes(BaseEstimator, TransformerMixin):
    
    def __init__(self, toRemove, toTransform):
        assert type(toRemove) == list
        
        self.toRemove = toRemove
        self.toTransform = toTransform
    
    def fit(self, X, y=None):
        self.X_df = pd.DataFrame(X, columns=brca_data.columns)
        
        self.mainlist = list(brca_data.columns)
        
        for attrib in self.toRemove:
            self.mainlist.remove(attrib)
                
        # The following block finds the desired number of zero decimal places for the positive minimum value of each feature.
        self.smallestNonZero = 0
        self.leastNum_per_feature = []
        
        for feature in self.mainlist:
            next_largest = max(self.X_df[feature])
            for x in self.X_df[feature]:
                if x < next_largest and x > 0:
                    next_largest = x
            self.smallestNonZero = 1/10**len(str(next_largest))
            self.leastNum_per_feature.append([feature, self.smallestNonZero])
            
        self.leastNum_per_feature = dict(self.leastNum_per_feature)
        
        return self
            
    def transform(self, X, y=None):
        check_is_fitted(self,'leastNum_per_feature')
        
        X = self.X_df.copy()
        
        if self.toTransform == True:
            for feature in self.mainlist:
                X.loc[X[feature]==0, [feature]] = self.leastNum_per_feature[feature]

        return X.values
```

```python
# The following transformer will normalize the columns of our dataframe.

class DistribNormalizer(BaseEstimator, TransformerMixin):
    
    def __init__(self, toRemove, skew_thresh, kurt_thresh, NormalAll):
        assert type(toRemove) == list
        assert type(NormalAll) == bool
        
        self.NormalAll = NormalAll
        self.toRemove = toRemove
        self.skew_thresh = skew_thresh
        self.kurt_thresh = kurt_thresh
        
    def fit(self, X, y=None):
        
        self.mainlist = list(brca_data.columns)
        self.X_df = pd.DataFrame(X, columns = list(brca_train.columns))
        
        for attrib in self.toRemove:
            self.mainlist.remove(attrib)
        
        # Create an iterable of features that from being normally distributed with respect to a given parameter.
        self.abnormal_dist = list()
        
        for diagnosis in [0, 1]:
            for feature in self.mainlist:
                if self.X_df.loc[self.X_df.loc[:,'diagnosis']==diagnosis, feature].skew() > self.skew_thresh:
                    self.abnormal_dist.append(feature)
                elif self.X_df.loc[self.X_df.loc[:,'diagnosis']==diagnosis, feature].skew() < -self.skew_thresh:
                    self.abnormal_dist.append(feature)
                elif self.X_df.loc[self.X_df.loc[:,'diagnosis']==diagnosis, feature].kurtosis() > self.kurt_thresh:
                    self.abnormal_dist.append(feature)
                elif self.X_df.loc[self.X_df.loc[:,'diagnosis']==diagnosis, feature].kurtosis() < -self.kurt_thresh:
                    self.abnormal_dist.append(feature)
        
        self.abnormal_set = set(self.abnormal_dist)
        
        return self
    
    def transform(self,X,y=None):
        brca_df = pd.DataFrame(X, columns=list(brca_train.columns))
        brca_df_diagnosis = brca_df.loc[:,'diagnosis']
        brca_df_feature = brca_df.drop(columns=self.toRemove).astype(float)
        brca_imputed_df = pd.concat([brca_df_diagnosis, brca_df_feature], axis=1)

        logged_B_dfcols_list = [brca_df.loc[brca_df['diagnosis']==0, 'diagnosis']]
        logged_M_dfcols_list = [brca_df.loc[brca_df['diagnosis']==1, 'diagnosis']]

        for diagnosis in [0,1]:
            for attrib in self.mainlist:
                if diagnosis == 0 and attrib not in self.abnormal_set and self.NormalAll ==False:
                    logged_B_dfcols_list.append(np.log10(brca_df.loc[brca_df['diagnosis']==diagnosis, attrib].astype(float)))
                if diagnosis == 1 and attrib not in self.abnormal_set and self.NormalAll == False:
                    logged_M_dfcols_list.append(np.log10(brca_df.loc[brca_df['diagnosis']==diagnosis, attrib].astype(float)))
                elif diagnosis == 0 and attrib not in self.abnormal_set and self.NormalAll == True:
                    logged_B_dfcols_list.append(np.log10(brca_df.loc[brca_df['diagnosis']==diagnosis, attrib].astype(float)))
                elif diagnosis == 1 and attrib not in self.abnormal_set and self.NormalAll == True:
                    logged_M_dfcols_list.append(np.log10(brca_df.loc[brca_df['diagnosis']==diagnosis, attrib].astype(float)))
                elif diagnosis == 0 and attrib in self.abnormal_set:
                    logged_B_dfcols_list.append(np.log10(brca_df.loc[brca_df['diagnosis']==diagnosis, attrib].astype(float)))
                elif diagnosis == 1 and attrib in self.abnormal_set:
                    logged_M_dfcols_list.append(np.log10(brca_df.loc[brca_df['diagnosis']==diagnosis, attrib].astype(float)))

        brca_B_gaussed_df = pd.concat(logged_B_dfcols_list, axis=1)
        brca_M_gaussed_df = pd.concat(logged_M_dfcols_list, axis=1)
        brca_gaussed_df = pd.concat([brca_B_gaussed_df,brca_M_gaussed_df]).sort_index()
        
        return brca_gaussed_df.values
```

```python
import statistics

class OutlierTreatment(BaseEstimator, TransformerMixin):
    
    def __init__(self, z_score, sortby, middleness):
        assert type(z_score) == int or list or dict
        assert type(sortby) == list
        assert type(middleness) == str
        
        self.z_score = z_score
        self.sortby = sortby
        self.middleness = middleness
                    
    def fit(self,X,y=None):
        
        self.Outlier_stats = dict()
        self.mainlist = dict([[feature, index+1] 
                              for index, feature in zip(range(0,len(list(np.array(brca_data.drop(columns=['diagnosis']).columns)))), 
                                                        list(np.array(brca_data.drop(columns=['diagnosis']).columns)))])
        self.X_df = pd.DataFrame(X, columns = list(brca_train.columns))
        self.X_features_df = self.X_df.drop(columns=['diagnosis']).astype(float)
        self.X_diagnosis_df = self.X_df['diagnosis']
        self.X_df = pd.concat([self.X_diagnosis_df, self.X_features_df],axis=1)
        
        self.means_df = self.X_df.groupby(self.sortby).agg(self.middleness).reset_index(drop=False)
        self.stdev_df = self.X_df.groupby(self.sortby).agg('std').reset_index(drop=False)
        
        for diagnosis in [0,1]:
            for feature in list(self.mainlist):
                if diagnosis == 0:
                    self.Outlier_stats[f"{feature}_{diagnosis}"] = {f'{diagnosis}_lower_outlier <':self.means_df.loc[0,feature] - self.z_score*self.stdev_df.loc[0,feature], 
                                                                    f'{diagnosis}_upper_outlier >':self.means_df.loc[0,feature] + self.z_score*self.stdev_df.loc[0,feature]}
                if diagnosis == 1:
                    self.Outlier_stats[f"{feature}_{diagnosis}"] = {f'{diagnosis}_lower_outlier <':self.means_df.loc[1,feature] - self.z_score*self.stdev_df.loc[1,feature], 
                                                                    f'{diagnosis}_upper_outlier >':self.means_df.loc[1,feature] + self.z_score*self.stdev_df.loc[1,feature]}
        
        return self
    
    def transform(self, X, y=None):
        self.outlier_set = set()
        self.outlier_rowIndices = []
        
        for diagnosis in [0,1]:
            for feature in list(self.mainlist):
                upperlimit = self.Outlier_stats[f"{feature}_{diagnosis}"][f'{diagnosis}_upper_outlier >']
                lowerlimit = self.Outlier_stats[f"{feature}_{diagnosis}"][f'{diagnosis}_lower_outlier <']
                
                bool_indices = self.X_df.index[self.X_df['diagnosis'] == diagnosis].tolist()
                upper_outliers = self.X_df.loc[bool_indices, feature].index[self.X_df.loc[bool_indices, feature] > upperlimit].tolist()
                lower_outliers = self.X_df.loc[bool_indices, feature].index[self.X_df.loc[bool_indices, feature] < lowerlimit].tolist()
                             
                self.outlier_rowIndices = self.outlier_rowIndices + upper_outliers + lower_outliers
                        
        self.outlier_set = set([list_entry for list_entry in self.outlier_rowIndices])
        
        return self.X_df.drop(self.X_df.index[list(self.outlier_set)]).values
```

We intend to use sklearn's ColumnTransformer class to transform the columns of our independent variables and encode the label columns. Since the ColumnTransformer class does this seperately and the column-transformation may remove some outliers, the dimensions of the array of independent variables may not match the 1-d array of the independent variable ('diagnosis'). Thus, we need to do one hot encoding together with the column transformation. However, one-hot-encoding the label-column makes the code marginally less-readable because OneHotEncoder returns a matrix which will force us to write 'more complicated' code for the other custom transformers. Hence, we constructed a simple customer transformer called the BinaryEncoder.

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

# Note sklearn pipeline ignores categorical features. Make sure to Hot-encode them.

standard_scaler = StandardScaler()

Transformation_pipeline = Pipeline([('BinaryEncoder', BinaryEncoder()),
                                    ('noZeroes', noZeroes(toRemove = ['diagnosis'], toTransform=True)),
                                    ('normalizer', DistribNormalizer(toRemove=['diagnosis'], skew_thresh=1, kurt_thresh=1, NormalAll=True)),
                                    ('outlier_remover', OutlierTreatment(z_score=4, sortby=['diagnosis'], middleness='mean')),
                                    ('feature_scaler', standard_scaler)])

toTransform_cols = list(brca_train.columns)

columnTransformer = ColumnTransformer([('column_transformer', Transformation_pipeline, toTransform_cols)])

preprocessed_data = columnTransformer.fit_transform(brca_train)

preprocessed_data
```

```python
# We now seperate the independent from the dependent variable to fit on our classifier.
brca_Xtrain = pd.DataFrame(preprocessed_data[:,1:], columns=brca_data.drop(columns=['diagnosis']).columns)
brca_Ytrain = preprocessed_data[:,[0]].ravel().astype(int)

brca_test_prepared = columnTransformer.fit_transform(brca_test)
brca_Xtest = pd.DataFrame(brca_test_prepared[:,1:])
brca_Ytest = brca_test_prepared[:,[0]].ravel().astype(int)

brca_Ytrain
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

LogisticRegressor = LogisticRegression()
LogisticRegressor.fit(brca_Xtrain, brca_Ytrain)

# We now test on our test set.
y_pred = LogisticRegressor.predict(brca_Xtest)

acc = accuracy_score(brca_Ytest, y_pred)
acc
```

```python
# To check for overfitting, we also evaluate model-performance on our train set. Drastically higher scores on the training set
# means we overfitted on the training data. We want similar scores.

ytrain_pred = LogisticRegressor.predict(brca_Xtrain)

train_set_acc = accuracy_score(brca_Ytrain, ytrain_pred)
train_set_acc
```

```python
# Performing cross-validation manually.
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

skfolds = StratifiedKFold(n_splits=3)

for train_index, test_index in skfolds.split(tumor_features, diagnosis):
    clone_transformer = clone(columnTransformer)
    clone_clf = clone(LogisticRegressor)
    
    brca_train_Xfolds = tumor_features.loc[train_index]
    brca_train_yfolds = diagnosis.loc[train_index]
    brca_test_Xfolds = tumor_features.loc[test_index]
    brca_test_yfolds = diagnosis.loc[test_index]
    
    training_data4fit = pd.concat([brca_train_yfolds, brca_train_Xfolds], axis=1)
    prepared_traindata = clone_transformer.fit_transform(training_data4fit)
    test_data4pred = pd.concat([brca_test_Xfolds, brca_test_yfolds], axis=1)
    prepared_testdata = clone_transformer.fit_transform(test_data4pred)
    
    clone_clf.fit(prepared_traindata[:,1:], prepared_traindata[:,[0]].ravel().astype(int))
    y_pred = clone_clf.predict(prepared_testdata[:,1:])
    print(accuracy_score(prepared_testdata[:,[0]].ravel().astype(int), y_pred))
```
